```python
import random 
import numpy as np 
import matplotlib.pyplot as plt 
#random.seed(7)

#Generating the random target function :
w_tar = [random.uniform(-1,1) for i in range(3)]

#Generating random data points :
size = 150
#size = int(input("Enter data size :"))
x1_data = [random.uniform(-1,1) for i in range(size)]
x2_data = [random.uniform(-1,1) for i in range(size)]
datapoints = []
for i in range(size) :
    datapoints.append([1,x1_data[i],x2_data[i]])

#Signum function to determine the sign of any dot product:
def SGN(pnt,w):
    val = np.dot(w,pnt) #Taking the dot product
    if val > 0 :
        return 1 
    elif val < 0 :
        return -1
    else :
        return 0 

#Generating the outputs for the data points from the target function :
y_tar = []
for i in range(size) :
    y_tar.append(SGN(datapoints[i],w_tar))

print(y_tar)
#Perceptron Learning Algorithm :
w = [0,0,0] #Weight vector
for n_iterations in range(1,10) :
    for j in range(size) :
        y_hyp = SGN(w, datapoints[j])
        if y_hyp != y_tar[j] :
            w += np.dot(y_tar[j], datapoints[j])
        print(f'weights_in_progress:{w}')
        
    x_g = [-w[0] / w[1], 0]
    y_g = [0, -w[0] / w[2]]

    #Plotting :
    fig, ax = plt.subplots() 
    ax.scatter(x1_data,x2_data,label="Data Points") 
    x1_t = [-w_tar[0]/w_tar[1] , 0]
    x2_t = [0, -w_tar[0]/w_tar[2]]
    ax.plot(x1_t,x2_t,color="red",label="Target Function")
    #for xy in zip(x1_data, x2_data):
    #plt.annotate('(%.2f, %.2f)' % xy, xy=xy)
    ax.plot(x_g,y_g,color="black",label="Perceptron")
    ax.set_title(f'Perceptron Learning Algorithm')
    ax.legend()

#Printing the results :

for i in range(size) :
    print(datapoints[i],"\t",y_tar[i])

print(f'Target function:{w_tar}')
print(f'Hypothesis weights:{w}')
```
